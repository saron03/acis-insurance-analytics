```python
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])
```

```python
monthly_summary = df.groupby('TransactionMonth')[['TotalClaims', 'TotalPremium']].sum().reset_index()
monthly_summary.head()
```

```python
plt.figure(figsize=(12,6))
plt.plot(monthly_summary['TransactionMonth'], monthly_summary['TotalClaims'], label='TotalClaims')
plt.plot(monthly_summary['TransactionMonth'], monthly_summary['TotalPremium'], label='TotalPremium')
plt.title('Total Claims and Total Premium Over Time')
plt.xlabel('Transaction Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)
plt.show()
```

```python
monthly_summary['LossRatio'] = monthly_summary['TotalClaims'] / monthly_summary['TotalPremium']
plt.figure(figsize=(12,6))
plt.plot(monthly_summary['TransactionMonth'], monthly_summary['LossRatio'], marker='o', color='purple')
plt.title('Monthly Loss Ratio Over Time')
plt.xlabel('Transaction Month')
plt.ylabel('Loss Ratio')
plt.grid(True)
plt.show()
```

```python
# Convert TransactionMonth to datetime if not already
df['TransactionMonth'] = pd.to_datetime(df['TransactionMonth'])

# Aggregate total claims and premiums by month
monthly_trends = df.groupby('TransactionMonth')[['TotalClaims', 'TotalPremium']].sum().reset_index()

# Plot trends over time
plt.figure(figsize=(14,6))
sns.lineplot(data=monthly_trends, x='TransactionMonth', y='TotalClaims', label='Total Claims')
sns.lineplot(data=monthly_trends, x='TransactionMonth', y='TotalPremium', label='Total Premium')
plt.title('Monthly Total Claims and Premiums Over Time')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Compute sums by Province
prov_summary = (
    df.groupby('Province')
      .agg(TotalClaims=('TotalClaims','sum'),
           TotalPremium=('TotalPremium','sum'))
      .assign(LossRatio=lambda x: x.TotalClaims / x.TotalPremium)
      .sort_values('LossRatio')
      .reset_index()
)

# 2. Display the table of loss ratios
print(prov_summary[['Province','LossRatio']])

# 3. Bar plot of Loss Ratio by Province
plt.figure(figsize=(10,6))
sns.barplot(data=prov_summary, x='LossRatio', y='Province',
            palette='viridis')
plt.title("Loss Ratio by Province")
plt.xlabel("Loss Ratio (TotalClaims / TotalPremium)")
plt.ylabel("Province")
plt.xlim(0, prov_summary.LossRatio.max()*1.1)
plt.show()
```
